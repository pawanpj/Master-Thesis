```python
import requests
temp = requests.get("https://zbw.eu/stw/version/latest/descriptor/about.de.html")
```

```python
content_parse =temp.content
```

```python
complete_content=[]
soup = BeautifulSoup(content_parse, 'html.parser')
```

```python
soup
```

```python
ZBW_Standard_Thesaurus_Wirtschaft = []
for li_tag in soup.find_all('li'):
    for j in li_tag:
        if (str(j).find("descriptor") > 0 and str(j).find("Alphabetische Liste der Deskriptoren") < 0):
            ZBW_Standard_Thesaurus_Wirtschaft.append(j.text.strip())
```

```python
import pickle

with open('ZBW_Standard_Thesaurus_Wirtschaft.pkl', 'wb') as f:
    pickle.dump(ZBW_Standard_Thesaurus_Wirtschaft, f)
```
