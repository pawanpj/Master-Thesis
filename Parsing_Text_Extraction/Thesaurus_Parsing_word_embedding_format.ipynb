```python
import requests
from bs4 import BeautifulSoup

vgm_url ='https://lod.gesis.org/thesoz/de/index'
html_text = requests.get(vgm_url).text
soup = BeautifulSoup(html_text, 'html.parser')

data =[]
for li_tag in soup.find_all('ul', {'class':'list-group alphabetical-search-results side-navi'}):
    for span_tag in li_tag.find_all('a'):
        if span_tag.find('span') is None:
            data.append((span_tag.text))
        else:
            for span_tag_full in span_tag.find_all('span', {'class':'sr-only'}):
                data.append(span_tag_full.text)
```

```python
data
```

```python
n = 17
final_data= []
del data[:n]
final_data.append(data)
```

```python
final_data
```

```python
final_data = [item for sublist in final_data for item in sublist]
```

```python
len(final_data)
```

```python
import re
key = []
value = []
point = []
for elem in list(final_data):
    if len(re.findall("Prefer using",elem)) != 0:
         key.append(elem.split("Prefer using")[0])
         value.append(elem.split("Prefer using")[1])
    else:
        point.append(elem)
        
       
```

```python
# Merge two lists
final_list = key + value + point
```
