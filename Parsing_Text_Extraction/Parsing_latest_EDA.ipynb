```python
# for single_question-  tag :<zofar:questionSingleChoice   : Answers if variable in response domain and label in answer option  
# + No answer for nested questions and only 1st question picked because of redundancy 
text_ques = []
text_ans = []
for i in range(len(soup.find_all('questionSingleChoice'))):
    text_ans.append([])
    check_ques = soup.find_all('questionSingleChoice')[i].find_all('question')
    if check_ques:
        if len(check_ques)>1:
            check_ques_len = len(check_ques[0].get_text().split())
            if check_ques_len!=0:
                temp_ques = " ".join(check_ques[0].get_text().split())
#                 if (temp_ques[0].isalpha()):
                text_ques.append(temp_ques)
                text_ans[i].append('')
            else:
                 text_ques.append('')

        else:
            check_ques = soup.find_all('questionSingleChoice')[i].find('question')
            if check_ques:
                check_ques_len = len(check_ques.get_text().split())
                if check_ques_len !=0:
                    temp_ques = " ".join(check_ques.get_text().split())
#                     if (temp_ques[0].isalpha()):#and ('#' not in temp_ques) :
                    text_ques.append(temp_ques)
                    temp_ans = soup.find_all('questionSingleChoice')[i].find('zofar:responseDomain')
                    if temp_ans.has_attr('variable'):
                        temp2 = temp_ans.find_all('answerOption')
                        for elem in temp2:
                            if 'label' in (elem.attrs.keys()):
                                text_ans[i].append(" ".join(elem['label'].split()))
                    else:
                        text_ans[i].append('')
                else:
                        text_ques.append('')
            else:
                    text_ques.append('')
    else:
                text_ques.append('')
#      else:
#         text_ques.append('')
```

```python
text_ques
```

```python
#adding full stop at the end of each sentence.
import string 
for i in range(len(text_ques)):
    if (len(text_ques[i])) > 1:
        if (text_ques[i][-1]) in string.punctuation:
            text_ques[i] = text_ques[i][:-1] + '.'
        else: 
            text_ques[i] = text_ques[i] + '.'       
```

```python
text_ques
```

```python
#adding full stop at the end of each answer option sentence.

for i in range(len(text_ans)):
    for j in range(len(text_ans[i])):
        if (len(text_ans[i][j])) > 1:
            if (text_ans[i][j][-1]) in string.punctuation:
                text_ans[i][j] = text_ans[i][j][:-1] + '.'
            else: 
                text_ans[i][j] = text_ans[i][j] + '.'
        
```

```python
# for single_question- Matrix - No answers taken # tag <zofar:matrixQuestionSingleChoice.

matrix_text_ques = []

for i in range(len(soup.find_all('matrixQuestionSingleChoice'))):
    print(i)

    check_ques = soup.find_all('matrixQuestionSingleChoice')[i].find_all('question')
    if check_ques:
        for elem in check_ques:
            check_ques_len = len(elem.get_text().split())
            if check_ques_len>1:
                temp_ques = " ".join(elem.get_text().split())
                if (temp_ques[0].isalpha()):
                    matrix_text_ques.append(temp_ques)
```

```python
#<zofar:matrixQuestionMixed - no answers taken

matrix_text_ques_mixed = []
for i in range(len(soup.find_all('matrixQuestionMixed'))):
    print(i)

    check_ques = soup.find_all('matrixQuestionMixed')[i].find_all('question')
    if check_ques:
        for elem in check_ques:
            check_ques_len = len(elem.get_text().split())
            if check_ques_len>1:
                temp_ques = " ".join(elem.get_text().split())
                if (temp_ques[0].isalpha()):
                    matrix_text_ques_mixed.append(temp_ques)
```

```python
#<zofar:matrixMultipleChoice
matrix_text_ques_multiple = []
for i in range(len(soup.find_all('matrixMultipleChoice'))):
    print(i)

    check_ques = soup.find_all('matrixMultipleChoice')[i].find_all('question')
    if check_ques:
        for elem in check_ques:
            check_ques_len = len(elem.get_text().split())
            if check_ques_len>1:
                temp_ques = " ".join(elem.get_text().split())
                if (temp_ques[0].isalpha()):
                    matrix_text_ques_multiple.append(temp_ques)
```

```python
#<zofar:matrixQuestionOpen - extract only questions
matrix_ques_open = []
for i in range(len(soup.find_all('matrixQuestionOpen'))):
    print(i)
    check_ques = soup.find_all('matrixQuestionOpen')[i].find_all('question')
    if check_ques:
        for elem in check_ques:
            check_ques_len = len(elem.get_text().split())
            if check_ques_len>1:
                temp_ques = " ".join(elem.get_text().split())
                if (temp_ques[0].isalpha()):
                    matrix_ques_open.append(temp_ques)
```

```python
#adding full stop at the end of each matrix_ques_open
import string 
for i in range(len(matrix_ques_open)):
    if (len(matrix_ques_open[i])) > 1:
        if (matrix_ques_open[i][-1]) in string.punctuation:
            matrix_ques_open[i] = matrix_ques_open[i][:-1] + '.'
        else: 
            matrix_ques_open[i] = matrix_ques_open[i] + '.' 
```

```python
# open ques(String) - No Answers taken
open_ques = []
for i in range(len(soup.find_all('questionOpen'))):
    if soup.find_all('questionOpen')[i].find('question'):
        check_ques_len = len(soup.find_all('questionOpen')[i].find('question').get_text().split())
        if check_ques_len!=0:
            temp_ques = " ".join(soup.find_all('questionOpen')[i].find('question').get_text().split())
            if (temp_ques[0].isalpha()): #and ('#' not in temp_ques) :
                open_ques.append(temp_ques)
```

```python
open_ques
```

```python
#adding full stop at the end of each matrix_ques_open
import string 
for i in range(len(open_ques)):
    if (len(open_ques[i])) > 1:
        if (open_ques[i][-1]) in string.punctuation:
            open_ques[i] = open_ques[i][:-1] + '.'
        else: 
            open_ques[i] = open_ques[i] + '.' 
```

```python
open_ques
```

```python
# For multipleChoice.
text_ques_new = []
text_ans_new = []
for i in range(len(soup.find_all('multipleChoice'))):
    print(i)
    text_ans_new.append([])
    text_ques_new.append([])
    check_ques = soup.find_all('multipleChoice')[i].find_all('question')
    for elm in check_ques:
        if elm:
            check_ques_len = len(elm.get_text().split())
            if check_ques_len !=0:
                temp_ques = " ".join(elm.get_text().split())
                text_ques_new[i].append(temp_ques)
                temp_ans = soup.find_all('multipleChoice')[i]
                temp2 = temp_ans.find_all('zofar:answerOption')
                for elem in temp2:
                    if 'variable'and 'label' in (elem.attrs.keys()):
                        text_ans_new[i].append(" ".join(elem['label'].split()))
                    else:
                        text_ans_new[i].append('')
            else:
                text_ques_new[i].append('')
        else:
            text_ques_new[i].append('')
```

```python
import string 
for i in range(len(text_ques_new)):
    if (len(text_ques_new[i])) > 1:
        if (text_ques_new[i][-1]) in string.punctuation:
            text_ques_new[i] = text_ques_new[i][:-1] + '.'
        else: 
            text_ques_new[i] = text_ques_new[i] + ['.'] 
```

```python
text_ques_new
```

```python
for i in range(len(text_ans_new)):
    for j in range(len(text_ans_new[i])):
        if (len(text_ans_new[i][j])) > 1:
            if (text_ans_new[i][j][-1]) in string.punctuation:
                text_ans_new[i][j] = text_ans_new[i][j][:-1] + '.'
            else: 
                text_ans_new[i][j] = text_ans_new[i][j] + '.'
```
